```python
from spotify_get_top_tracks import get_top_tracks
from genius_get_lyrics import get_lyrics_dictionary
from google_translate import get_translate
from ibm_emotion_analysis import emotion_analysis
import csv
import json
```

```python
# Initialize an empty list to store the values from the column
all_spotify_usernames = []

# Specify the file paths
file_paths = ['freshmen_spotify_usernames.csv', 'sophomores_spotify_usernames.csv', 'juniors_spotify_usernames.csv', 'seniors_spotify_usernames.csv'] #'freshmen_spotify_usernames.csv', 'sophomores_spotify_usernames.csv', 'seniors_spotify_usernames.csv'

for csv_file in file_paths:

    spotify_usernames = []

    # Open the CSV file
    with open(csv_file, 'r') as file:
        # Create a CSV reader object
        csv_reader = csv.reader(file)

        next(csv_reader)
        
        # Iterate over each row in the CSV file
        for row in csv_reader:
            # Append the value from the desired column to the list
            spotify_usernames.append(row[1])  # Assuming the column you want is the first column (index 0)

    # Print the list containing the column data
    all_spotify_usernames.append(spotify_usernames)
```

```python
amount_of_top_tracks = 2
title_index = 0

titles = ["freshmen", "sophomores", "juniors", 'seniors']

for class_usernames in all_spotify_usernames:

    class_top_tracks = get_top_tracks(class_usernames, amount_of_top_tracks)
    print(class_top_tracks)
    class_lyrics = get_lyrics_dictionary(class_usernames, class_top_tracks)
    print(class_lyrics)
    get_translate(class_lyrics)
    class_emotional_scores = emotion_analysis(class_lyrics)

    # save scores dictionary to a separate JSON file, one for each year
    with open(titles[title_index] + '_emotional_scores.json', 'w') as json_file:
        json.dump(class_emotional_scores, json_file)

    title_index += 1

# TODO: don't print anything
```

```python
emotions = ['sadness', 'joy', 'fear', 'disgust', 'anger']
year_labels = ['freshmen', 'sophomores', 'juniors', 'seniors']

scores_indiv = []
for year in year_labels:
    json_file = f'{year}_emotional_scores.json'
    with open(json_file, 'r') as file:
        scores_indiv.append(json.load(file))

# data aggregating; each year will have a variable of dictionary with emotion mapped to aggregate score
list_years = []
for year_label in year_labels:
    year_dict = {'sadness': 0, 'joy': 0, 'fear': 0, 'disgust': 0, 'anger': 0}
    list_years.append(year_dict)

for index in range(len(scores_indiv)):
    for _, emotion_scores in scores_indiv[index].items():
        for emotion, score in emotion_scores.items():
            list_years[index][emotion] = list_years[index][emotion] + score

# Get averaged scroes
list_averaged_scores = []
for year_data in list_years:
    total_sum = sum(year_data.values())
    averaged_scores = {emotion: value / total_sum for emotion, value in year_data.items()}
    list_averaged_scores.append(averaged_scores)

list_averaged_scores
```

```python
# Save the list to a JSON file
with open('list_averaged_scores.json', 'w') as json_file:
    json.dump(list_averaged_scores, json_file)
```
